```python
os.mkdir("./jsons")

fs_write = fsspec.filesystem("")

json_path = "./jsons/"
name = 1
out = scan_grib("./hrrr.t00z.wrfsfcf01.grib2")  
for i, message in enumerate(out): 
    with fs_write.open(json_path + str(name)+ ".json", "w") as f: 
        f.write(ujson.dumps(message))
        name += 1
```
