```python
from kerchunk.grib2 import build_idx_grib_mapping, parse_grib_idx
```

```python
# this will we applied to every dataframe
mapping = build_idx_grib_mapping("s3://noaa-gefs-pds/gefs.20170101/06/gec00.t06z.pgrb2af006", storage_options=dict(anon=True), remote_options=dict(anon=True))

idx_track = {}
for time in range(0, 24, 6):
    idx_track[time] = parse_grib_idx(f"s3://noaa-gefs-pds/gefs.20170101/{time:02}/gec00.t{time:02}z.pgrb2af006", storage_options=dict(anon=True))
```

```python
d0 = idx_track[0].merge(mapping.loc[~mapping["attrs"].duplicated(keep="first"), :], on="attrs", how="left", suffixes=("", "_mapping"))
d6 = idx_track[6].merge(mapping.loc[~mapping["attrs"].duplicated(keep="first"), :], on="attrs", how="left", suffixes=("", "_mapping"))
d12 = idx_track[12].merge(mapping.loc[~mapping["attrs"].duplicated(keep="first"), :], on="attrs", how="left", suffixes=("", "_mapping"))
d18 = idx_track[18].merge(mapping.loc[~mapping["attrs"].duplicated(keep="first"), :], on="attrs", how="left", suffixes=("", "_mapping"))
```
